For this next part you need to install the `geopandas` package. However, I had trouble installing this using `pip install geopandas` on my Windows system from the command terminal (I did not try conda). I kept getting this error:
```python
A GDAL API version must be specified. Provide a path to gdal-config using a GDAL_CONFIG environment variable or use a GDAL_VERSION environment variable.
```

If you run into this error, you can get around it by running the following four lines of code directly from your terminal:

```python
pip install pipwin
pipwin install gdal
pipwin install fiona
pip install geopandas
```

Solution sourced from Marcelo Machado Pereira's answer here: https://stackoverflow.com/questions/54734667/error-installing-geopandas-a-gdal-api-version-must-be-specified-in-anaconda

Also note that we need four types of mapping files to make this work: `xyz.shp`, `xyz.dbf`, `xyz.shx`, and `xyz.prj`. I downloaded the files here and placed them in the Dataset_2_House_Sale folder:
https://gis-kingcounty.opendata.arcgis.com/datasets/kingcounty::community-service-areas-with-subarea-details-in-king-county-community-service-detail-area/about

Additional details about these filetypes and how they're used can be found here: https://medium.com/@ianforrest11/graphing-latitudes-and-longitudes-on-a-map-bf64d5fca391

````python
import geopandas as gpd
from shapely.geometry import Point, Polygon
````

````python
street_map = gpd.read_file('./Dataset_2_House_Sales/Community_Service_Areas_with_Subarea_Details_in_King_County___community_service_detail_area.shp')
````

````python
# designate coordinate system
crs = {'init':'espc:4326'}

# zip x and y coordinates into single feature
geometry = [Point(xy) for xy in zip(house_main_df['long'], house_main_df['lat'])]

# create GeoPandas dataframe
geo_df = gpd.GeoDataFrame(house_main_df, geometry = geometry)

geo_df.head()
````

````python
fig, ax = plt.subplots(figsize=(12,12))
# add .shp mapfile to axes
street_map.plot(ax=ax, alpha=0.4,color='grey')
geo_df.plot(ax=ax,markersize=0.5)
````
